```python
# Tree Node Definition
class TreeNode:
    def _init_(self, val):
        self.val = val
        self.left = None
        self.right = None

# Breadth-First Search (BFS) on a Tree
def tree_bfs(root):
    if not root:
        return

    queue = [root]
    result = []

    while queue:
        current = queue.pop(0)
        result.append(current.val)

        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)

    return result

# Depth-First Search (DFS) on a Tree (Preorder)
def tree_dfs_preorder(root):
    if not root:
        return []

    result = []
    stack = [root]

    while stack:
        current = stack.pop()
        result.append(current.val)

        if current.right:
            stack.append(current.right)
        if current.left:
            stack.append(current.left)

    return result

# Depth-First Search (DFS) on a Tree (Inorder)
def tree_dfs_inorder(root):
    if not root:
        return []

    result = []
    stack = []
    current = root

    while current or stack:
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()
        result.append(current.val)
        current = current.right

    return result

# Depth-First Search (DFS) on a Tree (Postorder)
def tree_dfs_postorder(root):
    if not root:
        return []

    result = []
    stack1 = [root]
    stack2 = []

    while stack1:
        current = stack1.pop()
        stack2.append(current)

        if current.left:
            stack1.append(current.left)
        if current.right:
            stack1.append(current.right)

    while stack2:
        result.append(stack2.pop().val)

    return result
```

```python
import random
import time

def bfs(lis, goal):
    if not lis:
        return None

    queue = [(lis, 0)]
    visited = set()

    while queue:
        current, depth = queue.pop(0)
        visited.add(tuple(current))

        if current[-220:] == goal:
            return depth

        for i in range(len(current)):
            new_list = current[:i] + current[i+1:]
            if tuple(new_list) not in visited:
                queue.append((new_list, depth + 1))

    return -1

def dfs(lis, goal):
    if not lis:
        return None

    stack = [(lis, 0)]
    visited = set()

    while stack:
        current, depth = stack.pop()
        visited.add(tuple(current))

        if current[-220:] == goal:
            return depth

        for i in range(len(current)):
            new_list = current[:i] + current[i+1:]
            if tuple(new_list) not in visited:
                stack.append((new_list, depth + 1))

    return -1

# Example usage
random.seed(42)  # Set random seed for reproducibility

# Generate a random set of numbers
total_len = 1000
lis = [random.randint(1, 1000) for _ in range(total_len)]

goal = lis[total_len - 220:]

# Measure BFS execution time
start_time = time.time()
bfs_depth = bfs(lis, goal)
end_time = time.time()
execution_time = end_time - start_time
print("BFS Execution Time:", execution_time)
print("BFS Depth:", bfs_depth)

# Measure DFS execution time
start_time = time.time()
dfs_depth = dfs(lis, goal)
end_time = time.time()
execution_time = end_time - start_time
print("DFS Execution Time:", execution_time)
print("DFS Depth:", dfs_depth)
```
